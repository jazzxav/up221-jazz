```python
# We have to convert it do a data frame though using the reset.index
region_count = region_count.reset_index()
region_count
```

```python
# I want to make a bar graph to visually display the above table.
# Using the .plot.bar command, we can make a bar graph with an x and y axises and title
```

```python
region_count.plot.bar(x = 'Region', y = 'count', title = 'Number of Bikeways per LA Region')
```

```python
bike.info()
```

```python
bike[['Project_Name', 'Region', 'STR_FROM', 'STR_TO', 'ST_TYPE', 'CL_Miles', 'Lane_Miles', 'Year_', 'Shape__Length', 'geometry']]
```

```python
#to make a new dataset, we have to make a trimmed variable
trimmed_bike = bike[['Project_Name', 'Region', 'STR_FROM', 'STR_TO', 'ST_TYPE', 'CL_Miles', 'Lane_Miles', 'Year_', 'Shape__Length', 'geometry']].copy()
#now lets run the new variable
trimmed_bike
```

```python
#Using the plot command, we can now spatially display the trimmed data
trimmed_bike.plot()
```

```python
#Maybe we should make it look prettier
trimmed_bike.plot(
            figsize=(20,12),   #size of the plot (a bit bigger than the default)
            column = 'Region',   # Region column that defines the color of the dots
            legend = True,     # add a legend           
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1,1)
            }                  # this puts the legend to the side
) 
```
