# MAPS

```python
import geopandas as gpd
```

```python
# read in a geojson file downloaded from the LA Times
tracts=gpd.read_file('Data/Census_Tracts_2020.geojson')
tracts.head()
```

```python
#plot it
tracts.plot(figsize=(12,10))
```

```python
# we only really need FIPS and geometry, so let's subset the data
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
# create a FIPS column
tracts['FIPS'] ='06' + '037' + tracts['CT20']
```

```python
# check it!
tracts.head()
```

```python
# create a new dataframe based on the join
tracts_raceandtranstype=tracts.merge(df2,on="FIPS")
```

```python
tracts_raceandtranstype.head()
```

```python
tracts_raceandtranstype.plot(figsize=(12,10),
                 column='White Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_raceandtranstype.plot(figsize=(12,10),
                 column='Black or African American Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
# lets look at majority ethnic cluster in LA
# First, create new columns for percentages
tracts_raceandtranstype['PCT_Black'] = tracts_raceandtranstype['Black or African American Alone']/tracts_raceandtranstype['TotalPop']*100
tracts_raceandtranstype['PCT_White'] = tracts_raceandtranstype['White Alone']/tracts_raceandtranstype['TotalPop']*100
tracts_raceandtranstype['PCT_Asian'] = tracts_raceandtranstype['Asian Alone']/tracts_raceandtranstype['TotalPop']*100
```

```python
tracts_raceandtranstype.plot(figsize=(12,10),
                 column='PCT_Black',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_raceandtranstype[tracts_raceandtranstype.PCT_Asian > 5].plot(figsize=(12,10))
```

```python
#now lets import the folium to make the cool interactive map
import folium 
```

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 10,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts_raceandtranstype, # geo data
                  data=tracts_raceandtranstype, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS', 'PCT_White'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Percent Population White For Tracts With Highest Public Transit Commute (2022)').add_to(m)    # name on the legend color bar
m
```
